```julia
# Import packages.
using JuMP, SCS, Mayday, Mosek
```

```julia
m = Model(solver=SCSSolver())

@variable(m, γ)

a = generator(:a)
x = generator(:x)
y = generator(:y)
z = generator(:z)
v = defPolynomial(m, [:a, :x, :y, :z], 3)
w = defPolynomial(m, [:a, :x, :y, :z], 6)
r = defPolynomial(m, [:a, :x, :y, :z], 7)
s = defPolynomial(m, [:a, :x, :y, :z], 2)
p = a^8 + 2x^4 + 5y^4 - 2x^2 - 3y^2 - 3a^2 - 2z^2 + z^8 + 3
f = 4a^5 + x^3 - 4y^2 + 10z^5 - 1
g = a^2 + x - 5y - 3z^2
h = a + 2x + 14y - 13z
j = -3a^6 + x^2 - 12y^6 + 3z^2

@objective(m, Max, γ)

addSoSConstraint(m, p - γ + v*f + w*g + r*h + s*j)

m
```

```julia
solve(m)
```

```julia
getvalue(γ)
```

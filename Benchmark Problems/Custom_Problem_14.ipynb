```julia
# Import packages.
using JuMP, SCS, Mayday, Mosek
```

```julia
m = Model(solver=SCSSolver())

@variable(m, γ)

x = generator(:x)
y = generator(:y)
z = generator(:z)
v = defPolynomial(m, [:x, :y, :z], 3)
p = 2x^6 - x^4 + 3y^8 + 2z^4 + z^8 - 13
f = x^5 - 4y^3 + 10z^2 - 1

@objective(m, Max, γ)

addSoSConstraint(m, p - γ + v*f)

m
```

```julia
solve(m)
```

```julia
getvalue(γ)
```

```julia
# Import packages.
using JuMP, SCS, Mayday, Mosek
```

```julia
m = Model(solver=SCSSolver())

@variable(m, γ)

a = generator(:a)
b = generator(:b)
c = generator(:c)
x = generator(:x)
y = generator(:y)
v1 = defPolynomial(m, [:a, :b, :c, :x, :y], 3)
v2 = defPolynomial(m, [:a, :b, :c, :x, :y], 4)
v3 = defPolynomial(m, [:a, :b, :c, :x, :y], 5)
v4 = defPolynomial(m, [:a, :b, :c, :x, :y], 2)
v5 = defPolynomial(m, [:a, :b, :c, :x, :y], 1)
p = a^6 + + 4b^6 - 3b^2*c^3 + c^6 + 2x^4 + 5y^4 - 2x^2 - 3y^2 - 3a^2
f1 = 4a + b^2 + 7c^3 + x^3 - 4y^2 - 1
f2 = a^2 - 4b^2 + c + x - 5y
f3 = a + b - c +2x + 14y
f4 = -3a^4 + b^3 - c^4 + x^2 - 12y^4
f5 = 2a^5 - 8b^4 + 9c^5 + 2x^2 + 4y^2 

@objective(m, Max, γ)

addSoSConstraint(m, p - γ + v1*f1 + v2*f2 + v3*f3 + v4*f4 + v5*f5)

m
```

```julia
solve(m)
```

```julia
getvalue(γ)
```

```julia
# Import packages.
using JuMP, SCS, Mayday, Mosek
```

```julia
m = Model(solver=SCSSolver())

@variable(m, γ)

a = generator(:a)
b = generator(:b)
c = generator(:c)
x = generator(:x)
y = generator(:y)
z = generator(:z)
v = defPolynomial(m, [:a, :b, :c, :x, :y, :z], 3)
w = defPolynomial(m, [:a, :b, :c, :x, :y, :z], 4)
r = defPolynomial(m, [:a, :b, :c, :x, :y, :z], 5)
s = defPolynomial(m, [:a, :b, :c, :x, :y, :z], 2)
p = a^6 + + 4b^6 - 3b^2*c^3 + c^6 + 2x^4 + 5y^4 - 2x^2 - 3y^2 - 3a^2 - 2z^2 + z^6
f1 = 4a + b^2 + 7c^3 + x^3 - 4y^2 + 10z - 1
f2 = a^2 - 4b^2 + c + x - 5y - 3z^2
f3 = a + b - c +2x + 14y - 13z
f4 = -3a^4 + b^3 - c^4 + x^2 - 12y^4 + 3z^2

@objective(m, Max, γ)

addSoSConstraint(m, p - γ + v*f1 + w*f2 + r*f3 + s*f4)

m
```

```julia
solve(m)
```

```julia
getvalue(γ)
```
